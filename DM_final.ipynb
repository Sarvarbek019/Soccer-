```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
reader=pd.read_csv('stats.csv', delimiter=',',index_col=0);
reader
```

```python
reader["status"]=reader["league_position"].apply(lambda val:
                                            2 if val<=7 
                                                 else 1 if val<=17
                                              else 0)
```

```python
X=reader[['wins','wins_away','wins_home','draws','draws_home','draws_away','losses','losses_home','losses_away','points_per_game']]
```

# Data Models

### LinearRegression model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X=data.drop(['status'],axis=1)
Y=data['status']
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.coef_)
```

```python
regr = LinearRegression() 
regr.fit(X_train, y_train) 
print(regr.score(X_test, y_test))
```

## KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
yhat=knn.predict(X_test)
print(accuracy_score(yhat,y_test))
```

## SVC

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, svc.predict(X_test)))
```

## RFE with random forests

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
print(rf.feature_importances_)
```

```python
print(sum(rf.feature_importances_))
```

```python
mask = rf.feature_importances_> 0.1
print(mask)
```

```python
X_reduced = X.loc[:, mask]
print(X_reduced.columns)
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(estimator=RandomForestClassifier(),n_features_to_select=3, verbose=1)
rfe.fit(X_train,y_train)
```

```python
print(accuracy_score(y_test, rfe.predict(X_test)))
```
